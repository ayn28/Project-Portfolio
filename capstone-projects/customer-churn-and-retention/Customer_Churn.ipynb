## Dataset

### Data Overview

The dataset used in this project is available on Kaggle, [here.](https://www.kaggle.com/blastchar/telco-customer-churn)

Each row represents a customer and each column contains customers' attributes.
<br>**The dataset includes information about:**
* Demographic info about customers – gender, age range, and if they have partners and dependents
* Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies
* Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly charges, and total charges
* Customers who left within the last month - the column is called Churn

### Feature Overview

#### A. Demographics
>**gender** : Sex (categorical: "Male", "Female")
<br>**SeniorCitizen** : Is senior citizen (categorical: 0=no, 1=yes)
<br>**Partner** : Does customer have a partner (categorical: 0=no, 1=yes)
<br>**Dependents** : Does customer have dependents (categorical: 0=no, 1=yes)
#### B. Services
>**PhoneService** : Has phone service (categorical: "Yes", "No")
<br>**MultipleLines** : Has multiple lines (categorical: "Yes", "No", "No phone service")
<br>**InternetService** : Type of Internet service (categorical: "DSL", "Fiber optic", "No")
<br>**OnlineSecurity** : Has malware protection (categorical: "Yes", "No", "No internet service")
<br>**OnlineBackup** : Has digital backup service (categorical: "Yes", "No", "No internet service")
<br>**DeviceProtection** : Has device protection plan (categorical: "Yes", "No", "No internet service")
<br>**TechSupport** : Has contacted tech support (categorical: "Yes", "No", "No internet service")
<br>**StreamingTV** : Has TV streaming service (categorical: "Yes", "No", "No internet service")
<br>**StreamingMovies** : Has movie streaming service (categorical: "Yes", "No", "No internet service")
#### C. Account Information
>**customerID** : customer identification number (categorical)
<br>**tenure** : Number of months the consumer has been a customer (numeric: 1,2,3,...72)
<br>**Contract** : Type of contact (categorical: "Month-to-month", "One year", "Two year")
<br>**PaperlessBilling** : Customer is billed via email (categorical: "Yes", "No")
<br>**PaymentMethod** : Method of payment on file (categorical: "Electronic check", "Mailed check", "Bank transfer (automatic)", "Credit card (automatic)")
<br>**MonthlyCharges** : Monthly fee (numeric)
<br>**TotalCharges** : Sum of all fees (numeric)
#### D. Target Variable
>**Churn** : Customers who left within the last month (categorical: "Yes", "No")
