The data used in this analysis are the **Team Per Game** stats, rather than the Opponent Per Game stats, for the regular NBA seasons from 2007 to 2017.

<br><br>
The **Team Per Game** stats are an average of a given team's statistics per game for the entirety of a season. 
<br>The **Opponent Per Game** stats are an average of an opposing team's statistics per game for the entirety of a season.

```python
#Load data separately into DataFrames for further modification later.
filenames = glob('*.txt')
df_list = [pd.read_csv(df) for df in sorted(filenames)]
```

```python
#Drop the Rank(Rk), Games(G), and Total Rebounds (TRB) columns from all DataFrames.
for df in df_list:
    df.drop(['Rk', 'G', 'TRB'], axis=1, inplace=True)
```
